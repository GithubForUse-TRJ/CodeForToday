```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.neural_network import MLPClassifier
```

```python
df=pd.read_excel('default of credit card clients (1).xlsx')
```

```python
df.head(2)
# df
```

```python
Y=df['default payment next month']
print(Y)
X=df.drop(['default payment next month','ID'],axis=1,inplace=False)
X.head()
```

```python
Model=MLPClassifier(solver='sgd',hidden_layer_sizes=(7,6,4),random_state=1, learning_rate_init=0.00000004,learning_rate='adaptive', verbose=True, validation_fraction=0.1, early_stopping=False, max_iter=500)
```

```python
Model.fit(X,Y)
```

```python
plt.plot(Model.loss_curve_)
```

```python
Model.score(X,Y)
```
