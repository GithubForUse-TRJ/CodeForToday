```python
# Practical 2.2----------------
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
bstfeatures=SelectKBest(score_func=chi2, k=10)
fit=bstfeatures.fit(x,y)
dfscr=pd.DataFrame(fit.scores_)
dfcol=pd.DataFrame(x.columns)
```

```python
featureScores=pd.concat([dfcol,dfscr],axis=1)
featureScores.columns=['Name of Column', 'Score of Column']
print(featureScores.nlargest(100,'Score of Column'))
```
