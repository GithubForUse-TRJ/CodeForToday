```python
import pandas as pd
from sklearn.cluster import KMeans
```

```python
df=pd.read_csv('data.csv')
df.head()
```

```python
find=['id','diagnosis','Unnamed: 32']
df1=df.drop(find, inplace=False,axis=1)
df1.head()
```

```python
kmeans=KMeans(n_clusters=2,random_state=0).fit(df1)
kmeans.labels_
```
